```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_excel("Telco_customer_churn.xlsx")
```

```python
df.shape
df.info()
df.describe()
df.isnull().sum()
```

```python
df['Churn Label'].value_counts()

sns.countplot(x='Churn Label', data=df)
plt.title("Customer Churn Distribution")
plt.show()
```

## Insight: Shows how many customers churned vs stayed

## Churn by Contract Type

```python
sns.countplot(x='Contract', hue='Churn Label', data=df)
plt.title("Churn by Contract Type")
plt.xticks(rotation=45)
plt.show()
```

- Month-to-month customers usually churn more.

## Tenure vs churn

```python
sns.boxplot(x='Churn Label', y='Tenure Months', data=df)
plt.title("Tenure vs Churn")
plt.show()
```

- Short-tenure customers are more likely to churn

## Monthly charges vs churn

```python
sns.boxplot(x='Churn Label', y='Monthly Charges', data=df)
plt.title("Monthly Charges vs Churn")
plt.show()
```

- Higher monthly charges often lead to churn.

## Services Impact on Churn

```python
sns.countplot(x='Tech Support', hue='Churn Label', data=df)
plt.title("Churn by Tech Support")
plt.show()
```

- Customers without tech support churn more.

## Payment Method Analysis

```python
sns.countplot(x='Payment Method', hue='Churn Label', data=df)
plt.xticks(rotation=45)
plt.title("Churn by Payment Method")
plt.show()
```

## Churn Reason Analysis

```python
df['Churn Reason'].value_counts().head(10).plot(kind='bar')
plt.title("Top Reasons for Customer Churn")
plt.show()
```

- Competition and pricing are major churn drivers.

## Key Insights
- Month-to-month contracts have the highest churn.
- Customers with short tenure are more likely to leave.
- Higher monthly charges increase churn probability.
- Lack of tech support and security services correlates with churn.
- Competitor-related reasons dominate churn causes.

## Conclusion
- This EDA reveals that customer churn is strongly influenced by:
- Contract type
- Tenure duration
- Monthly charges
- Service quality
### The company should focus on:
- Encouraging long-term contracts
- Offering discounts to high-risk customers
- Improving service bundles
